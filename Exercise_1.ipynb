**Exercise 11: Write a program to display all prime numbers within a range**

image.png


```python
start = int(input())
end = int(input())

print ('Prime numbers between {} and {} are: '.format(start, end))

while start <= end :

  prime= 0; 
  for i in range( 2,start//2) :
    if start % i == 0 :
      prime=1 
      break 

  if not prime : print (start)

  start +=1
```
