**Exercise 6: Create a recursive function**


---

image.png

```python
def rec(end):
    if (end ==0):
        sum = 0
        return sum 
    else: 
        sum = end + rec(end-1)
        return sum 


print(rec(10))
```

**Exercise 8: Generate a Python list of all the even 
numbers between 4 to 30**


---


image.png

```python
list =[]

def even(satrt , end ):
    for i in range(satrt ,end):
      if i%2==0 :
        list.append(i)
    return list 

print (even(4, 30))
```
